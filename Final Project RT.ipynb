```python
#Every package we have ever run in python "just in case"

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn

from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
```

```python
#Accuracy Score

mag=pd.read_csv("SDS_Mag.csv",index_col=0)

X = mag
y = mag.cls
mag_class = ["Hadron", "Gamma"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
```

```python
#Estimating K-Groups
K_group = np.arange(1,18)
Train_acc = np.empty(len(K_group))
Test_acc = np.empty(len(K_group))

for i, k in enumerate(K_group):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    Train_acc[i] = knn.score(X_train, y_train)
    Test_acc[i] = knn.score(X_test, y_test)

plt.plot(K_group, Test_acc, label = 'Testing')
plt.plot(K_group, Train_acc, label = 'Training')
 
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
#Confusion Matrix

K_class = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
disp = plot_confusion_matrix(K_class, X_test, y_test, display_labels = mag_class)
plt.title("kNN Classifier")

print(disp.confusion_matrix)
```
